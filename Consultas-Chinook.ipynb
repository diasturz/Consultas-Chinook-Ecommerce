24. Forneça uma consulta que mostra a faixa mais comprada de 2013.

```python
conn = sql.connect("C:/Users/artur/Downloads/chinook/chinook.db")

df1 = pd.read_sql_query("""SELECT Name AS NameMusic,TrackId
                            FROM tracks""",conn)
df2 = pd.read_sql_query("""SELECT TrackId,InvoiceId
                            FROM invoice_items
""",conn)

df3 = pd.read_sql_query("""SELECT InvoiceId,InvoiceDate
                            FROM invoices
                            WHERE strftime('%Y', InvoiceDate) = '2013'""",conn)

conn.close()

df4 = pd.merge(df3,df2,how='left',on='InvoiceId')
df5 = pd.merge(df4,df1,how='left',on='TrackId')
df5.NameMusic.value_counts().sort_values(ascending=False).index[0]
```

26. Forneça uma consulta que mostra os top 3 artistas mais vendidos de toda base de dados.

```python
conn = sql.connect("C:/Users/artur/Downloads/chinook/chinook.db")

dfartist= pd.read_sql_query("""SELECT *
                            FROM artists""",conn)
dfartist


dfalbum = pd.read_sql_query("""SELECT AlbumId,ArtistId
                            FROM albums""",conn)
dfalbum


dftracks = pd.read_sql_query("""SELECT AlbumId,TrackId
                            FROM tracks""",conn)
dftracks

dfinvoiceitems = pd.read_sql_query("""SELECT TrackId,InvoiceId
                            FROM invoice_items""",conn)
dfinvoiceitems

dfinvoices = pd.read_sql_query("""SELECT InvoiceId
                            FROM invoices""",conn)

conn.close()

df1 = pd.merge(dfinvoices,dfinvoiceitems,how='left',on='InvoiceId')
df2 = pd.merge(df1,dftracks,how='left',on='TrackId')
df3 = pd.merge(df2,dfalbum,how='left',on='AlbumId')
df4 = pd.merge(df3,dfartist,how='left',on='ArtistId')
df4.Name.value_counts().nlargest(3)
```
